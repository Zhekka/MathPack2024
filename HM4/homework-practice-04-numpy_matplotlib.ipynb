```python
# 1. равномерное разбиение интервала
import numpy as np

def uniform_division_of_interval(a, b, n):
    return np.linspace(a, b, n + 1)

print(uniform_division_of_interval(-1.3, 2.5, 64))
```

```python
# 2. numpy массив длины 3n, заполненный циклическими числами 1, 2, 3, 1, 2, 3, 1, ...
def array_with_cyclic_num(n):
    arr = np.empty(3 * n, dtype=int)
    for i in range(3 * n):
        arr[i] = (i % 3) + 1
    return arr
    
print(array_with_cyclic_num(10))
```

```python
# 3. массив первых 10 нечетных целых чисел.
def array_with_odd_num(n):
    return np.arange(1, 2 * n + 1, 2)
    
print(array_with_odd_num(10))
```

```python
# 4. массив нулей размера 10 x 10, а затем создайте в нём "рамку" из единиц по краям
def single_frame(n):
    arr = np.zeros((n, n), dtype=int)
    arr[0, :] = 1
    arr[:, 0] = 1
    arr[-1, :] = 1
    arr[:, -1] = 1
    return arr 
    
print(single_frame(10))
```

```python
# 5. массив 8 x 8 с шахматной доской из нулей и единиц
def chess_board(n):
    arr = np.zeros((n, n), dtype=int)
    arr[::2, ::2] = 1
    arr[1::2, 1::2] = 1
    return arr 
    
print(chess_board(8))
```

```python
# 6. функция, которая создает n×n матрицу с (i,j)-элементом, равным i+j
def matrix(n):
    return np.array([[i + j for j in range(n)] for i in range(n)])

print(matrix(5))
```

```python
# 1. График x^2 * sin(1/x^2) + x на интервале [-1,1], используя 250 точек
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 250)
func = x**2 * np.sin(1 / x**2) + x

plt.plot(x, func)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$x^2 \\cdot \\sin(1/x^2) + x$')
plt.grid(True)
plt.show()
```

```python
# 2. semilogy график относительной разницы функций 1/(1 + x^2) и 1/x^2 на интервале [5,25]
x = np.linspace(5, 25, 100)
func1 = 1 / (1 + x**2)
func2 = 1 / x**2
rel_diff = np.abs(1 - func1 / func2)

plt.semilogy(x, rel_diff)
plt.xlabel('x')
plt.ylabel('Относительная разница')
plt.title('Относительная разница между $\\frac{1}{1 + x^2}$ и $\\frac{1}{x^2}$')
plt.grid(True)
plt.show()
```

```python
# 3. функции из scipy: гамма-функция (gamma) и гипергеометрическая функция (hyp2f1)
from scipy.special import gamma, hyp2f1

x = np.linspace(0, 1)
func1 = gamma(x)
func2 = hyp2f1(2, 5, 7, x)

plt.plot(x, func1, label = "gamma")
plt.plot(x, func2, label = "hyp2f1")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Гамма-функция и гипергеометрические функция')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. цветовой график sin(x)sin(y) на интервале [-pi, pi] x [-pi, pi]
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
func = np.sin(X) * np.sin(Y)

plt.figure()
plt.pcolor(X, Y, func, cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.title('Цветовой график $\sin(x) \sin(y)$')
plt.show()
```

```python
def f(x):
    return (2*x - 1) / (2*x + 1)

def rectangle_method(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    return dx * np.sum(f(x[:-1]))


def trapezoidal_method(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    return dx * (0.5 * f(a) + 0.5 * f(b) + np.sum(f(x[1:-1])))


def simpson_method(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    even_sum = np.sum(f(x[2:-1:2]))
    odd_sum = np.sum(f(x[1:-1:2]))
    return dx / 3 * (f(a) + f(b) + 2 * even_sum + 4 * odd_sum)


a = 0
b = 2
n = 10

integral_prev_rectangular = rectangle_method(f, a, b, n)
integral_prev_trapezoidal = trapezoidal_method(f, a, b, n)
integral_prev_simpsons = simpson_method(f, a, b, n)

while True:
    n *= 10
    integral_curr_rectangular = rectangle_method(f, a, b, n)
    integral_curr_trapezoidal = trapezoidal_method(f, a, b, n)
    integral_curr_simpsons = simpson_method(f, a, b, n)

    if abs(integral_curr_rectangular - integral_prev_rectangular) < 1e-5 and abs(
            integral_curr_trapezoidal - integral_prev_trapezoidal) < 1e-5 and abs(
            integral_curr_simpsons - integral_prev_simpsons) < 1e-5:
        break

    integral_prev_rectangular = integral_curr_rectangular
    integral_prev_trapezoidal = integral_curr_trapezoidal
    integral_prev_simpsons = integral_curr_simpsons

rectangular_integral = integral_curr_rectangular
trapezoidal_integral = integral_curr_trapezoidal
simpsons_integral = integral_curr_simpsons

print(f'Метод прямоугольников: {rectangular_integral} \nМетод трапеций: {trapezoidal_integral} \nПо правилу Симпсона: {simpsons_integral}')
```
